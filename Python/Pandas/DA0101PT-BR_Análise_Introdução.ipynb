<h2>Lendo Dados</h2>

Como os dados não possuem cabeçalhos, podemos adicionar um argumento <code>headers = None</code> dentro do método <code>read_csv()</code> para que o pandas não defina automaticamente a primeira linha como cabeçalho. <br>

Você também pode atribuir o caminho de dados a qualquer váriavel que você criar.

<h3>Adicionando Cabeçalho</h3>
<p>
Dê uma olhada em nosso conjunto de dados. O Pandas automaticamente define os cabeçalhos com número inteiro começando pelo 0.
</p>
<p>
Para melhor descrever nossoas dados, podemos adicionar cabeçalhos. Esta informação está disponível em:  <a href="https://archive.ics.uci.edu/ml/datasets/Automobile?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDA0101ENSkillsNetwork20235326-2022-01-01" target="_blank">https://archive.ics.uci.edu/ml/datasets/Automobile</a>.
</p>
<p>
Assim, temos que adicionar os cabeçalhos manualmente.
</p>
<p>
Primeiro, criamos uma lista "headers" que inclui todas as colunas ordenadas por nome.
Em seguida, usamos <code>dataframe.columns = headers</code> para definir os cabeçalhos conforme a lista que criamos.
</p>


````python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
````

````python
df.columns = headers
df.head(10)
````

Precisamos susbtituir o simbolo "?" por NaN para que <code> dropna()</code> possa remover os valores nulos.

````python
df1=df.replace('?',np.NaN)
df1.head(5)
````

Podemos descartar os valores vazios ao longo da coluna "price" da seguinte maneira:

````python
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
````

Agora, lemos com sucesso o conjunto de dados brutos e adicionamos os cabeçalhos corretos ao dataframe.

<h2>Salvando um conjunto de dados</h2>
<p>
Da mesma forma, o Pandas nos permite salvar o conjunto de dados em cvs. Ao usar o método <code>dataframe.to_csv()</code>, você pode adicionar um caminho e o nome do arquivo junto com as aspas entre parênteses.
</p>
<p>
Por exemplo, se você salvar o dataframe <b>df</b> como <b>automobile.csv</b> na sua máquina local, você poderá usar a sintaxe abaixo, onde <code>index = False</code> isso significa que o nome das linhas não serão gravados.
</p>

Também podemos ler e salvar outros formatos de arquivos. Podemos usar funções similares, como **`pd.read_csv()`** e **`df.to_csv()`** para outros dados de formatos. As funções são listadas na tabela a seguir:

<h2>Read/Save Other Data Formats</h2>

| Data Formate |        Read       |            Save |
| ------------ | :---------------: | --------------: |
| csv          |  `pd.read_csv()`  |   `df.to_csv()` |
| json         |  `pd.read_json()` |  `df.to_json()` |

| hdf          |  `pd.read_hdf()`  |   `df.to_hdf()` |
| sql          |  `pd.read_sql()`  |   `df.to_sql()` |
| ...          |        ...        |             ... |

<h2>Tipo de dados</h2>
<p>
Os dados têm uma variedade de tipos.<br>

Os principais tipos armazenados nos dataframes do Pandas são <b>object</b>, <b>float</b>, <b>int</b>, <b>bool</b> and <b>datetime64</b>. Para aprendermos melhor sobre cada atributo, é sempre bom para sabermos o tipo de dado de cada coluna. Em Pandas:
</p>

````python
df.dtypes
````

Uma série com o tipo de dados de cada coluna é retornada.

Como mostrado acima, é claro que o tipo de dados de "symboling" e "curb-weight" são <code>int64</code>, "normalized-losses" é <code>object</code> e "wheel-base" é <code>float64</code>, etc.

Esses tipos de dados podem ser alterados; aprenderemos como fazer isso em um módulo posterior.

<h1>Describe</h1>
Se quisermos ter um resumo estatistico de cada coluna, por exemplo, contagem, valor médio da coluna, desvio padrão da coluna, etc. Usamos o método <code>describe</code>


Este método fornecerá várias estatísticas resumidas, excluindo valores <code>NaN</code> (Não é um número).

````python
df.describe()
````

<p>
Isso mostra o resumo estatístico de todas as colunas de tipo numérico (int, float).<br>

Por exemplo, o atributo "simbolização" tem 201 contagens, o valor médio desta coluna é 0,84, o desvio padrão é 1,25, o valor mínimo é -2, 25º percentil é 0, 50º percentil é 1, 75º percentil é 2 e o valor máximo é 3.  <br>

No entanto, e se também quisermos verificar todas as colunas, incluindo aquelas que são do tipo objeto? <br><br>

Você pode adicionar um argumento <code>include = "all"</code> dentro do parenteses. Vamos tentar de novo.
</p

````python
# describe all the columns in "df" 
df.describe(include = "all")
````

<p>
Agora ele fornece o resumo estatístico de todas as colunas, incluindo atributos de tipo objeto.<br>

Agora podemos ver quantos valores únicos existem, qual é o valor superior e a frequência do valor superior nas colunas de tipo de objeto.<br>

Alguns valores na tabela acima mostram valores <code>NaN</code>. Isso ocorre porque esses números não estão disponíveis em relação a um determinado tipo de coluna.<br>

</p>
